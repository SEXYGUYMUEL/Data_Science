## 2) 데이터 증식
사용되어지는 데이터의 수가 부족할때 데이터 증식을 하기도 합니다.
이번 과제에서는 MNIST Data를 활용하여 데이터 증식을 하는 것을 목표로 합니다.

```python
## Data Import

from sklearn.datasets import fetch_openml
from scipy.ndimage.interpolation import shift
import matplotlib as mpl

mnist = fetch_openml('mnist_784', version=1, as_frame=False)
mnist.keys()
```

```python
#5 mnist가 어떤 data인지 시각화 하는 것이 제일 먼저입니다.
# data의 0번째 글자를 시각화 해주세요.
#your code here

X=mnist["data"]
some_digit = X[0]
some_digit_image = some_digit.reshape(28, 28)
plt.imshow(some_digit_image,cmap=mpl.cm.binary,interpolation="nearest")
plt.axis("off")
plt.show()
```

```python
#6 주어진 data를 shift 해서 데이터를 증식해 봅시다.
#your code here

def shift_image(image, dx, dy):
    image = image.reshape((28, 28))
    shifted_image = shift(image,[dy,dx],cval=0,mode='constant')
    return shifted_image.reshape([-1])
```

```python
#7 Shift한 data를 시각화해주세요.

image = X[0]
shifted_image_down = shift_image(image, 0, 5)
shifted_image_left = shift_image(image, -5, 0)

#your code here
plt.figure(figsize=(12,3))
plt.subplot(131)
plt.title("Original", fontsize=14)
plt.imshow(image.reshape(28,28),interpolation='nearest',cmap='Greys')
plt.subplot(132)
plt.title("Shifted down", fontsize=14)
plt.imshow(shifted_image_down.reshape(28,28),interpolation='nearest',cmap='Greys')
plt.subplot(133)
plt.title("Shifted left", fontsize=14)
plt.imshow(shifted_image_left.reshape(28,28),interpolation='nearest',cmap='Greys')
plt.show()
```

```python
#8 이번에는 주어진 이미지를 회전시켜 데이터 증식을 해보겠습니다.
# ccw는 반시계방향, cw는 시계방향을 뜻합니다.
#your code here
from scipy.ndimage.interpolation import rotate
image = X[0].reshape((28,28))
rotated_image_ccw = rotate(image,angle=90).reshape([-1])
rotated_image_cw = rotate(image,angle=-90).reshape([-1])

plt.figure(figsize=(12,3))
plt.subplot(131)
plt.title("Original", fontsize=14)
plt.imshow(image.reshape(28,28),interpolation='nearest',cmap='Greys')
plt.subplot(132)
plt.title("Rotated ccw", fontsize=14)
plt.imshow(rotated_image_ccw.reshape(28,28),interpolation='nearest',cmap='Greys')
plt.subplot(133)
plt.title("Rotated cw", fontsize=14)
plt.imshow(rotated_image_cw.reshape(28,28),interpolation='nearest',cmap='Greys')
plt.show()
```
