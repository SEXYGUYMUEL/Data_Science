```python
## 사용할 데이터를 import 합니다.

import numpy as np
import pandas as pd
import os


DOWNLOAD_ROOT = "https://raw.githubusercontent.com/rickiepark/handson-ml2/master/"
GDP_PATH = DOWNLOAD_ROOT + "datasets/lifesat/gdp_per_capita.csv"

gdp = pd.read_csv(GDP_PATH, thousands = ',', delimiter = '\t',
                  encoding = 'latin1', na_values = "n/a", index_col = "Country")

gdp.rename(columns = {"2015": "GDP per capita"}, inplace = True)
```

```python
#1 데이터가 어떤 형태인지 확인하기 위해 제일 위의 행 5개와 아래의 행 5개를 출력해주세요
# your code here
gdp
```

```python
#2 마지막 행은 불필요한 정보네요. 마지막 행 전체는 삭제해주세요. 삭제 후 마지막 5행을 출력해주세요.
# your code here

gdp=gdp.drop('International Monetary Fund, World Economic Outlook Database, April 2016')
gdp.tail(5)
```

```python
#3 GDP per capita의 값이 최대인 것과 최소인 것을 출력해주세요
# your code here

min_data = round(gdp["GDP per capita"].min(),3)
max_data = round(gdp["GDP per capita"].max(),3)

print("Minimum:",min_data, "\nMaximum:", max_data)
```

```python
#4 GDP per capita가 최소인 국가와 최대인 국가를 출력해주세요
# your code here

min_country = gdp.index[gdp["GDP per capita"]==gdp["GDP per capita"].min()][0]
max_country = gdp.index[gdp["GDP per capita"]==gdp["GDP per capita"].max()][0]

print("Minimum Country:",min_country, "\nMaximum Country:", max_country)
```
